```python
def parse_xy(s):
    string=s.Filesystem
    elements=string.split('/')[-1].split('-')
    x=int(elements[1][1:])
    y=int(elements[2][1:])
    s['x']=x
    s['y']=y
    return s
```

```python
def move_space(df,direc,sl):
    if direc=='up':
        t1=df.loc[sl[0],sl[1]]
        t2=df.loc[sl[0]-1,sl[1]]
        df.loc[sl[0],sl[1]]=t2
        df.loc[sl[0]-1,sl[1]]=t1
        sl=(sl[0]-1,sl[1])
    if direc=='down':
        t1=df.loc[sl[0],sl[1]]
        t2=df.loc[sl[0]+1,sl[1]]
        df.loc[sl[0],sl[1]]=t2
        df.loc[sl[0]+1,sl[1]]=t1
        sl=(sl[0]+1,sl[1])
    if direc=='left':
        t1=df.loc[sl[0],sl[1]]
        t2=df.loc[sl[0],sl[1]-1]
        df.loc[sl[0],sl[1]]=t2
        df.loc[sl[0],sl[1]-1]=t1
        sl=(sl[0],sl[1]-1)
    if direc=='right':
        t1=df.loc[sl[0],sl[1]]
        t2=df.loc[sl[0],sl[1]+1]
        df.loc[sl[0],sl[1]]=t2
        df.loc[sl[0],sl[1]+1]=t1
        sl=(sl[0],sl[1]+1)
    return sl
```
