```python
import pandas as pd
```

```python
df=pd.read_csv('Day22.csv')
```

```python
def format_data(s):
    s['Size']=int(s['Size'][:-1])
    s['Used']=int(s['Used'][:-1])
    s['Avail']=int(s['Avail'][:-1])
    s['Use%']=int(s['Use%'][:-1])
    return s
```

```python
def parse_xy(s):
    string=s.Filesystem
    elements=string.split('/')[-1].split('-')
    x=int(elements[1][1:])
    y=int(elements[2][1:])
    s['x']=x
    s['y']=y
    return s
```

```python
df2=df.apply(format_data,axis=1).apply(parse_xy,axis=1)
```

```python
dftest=df2[850:870]
```

```python
#df2[df2['Avail']>60]
dftest
```

```python
viable_pairs=0
for i,A in df2.iterrows():

    for j,B in df2.iterrows():
        if i!=j and A['Used']!=0 and B['Avail']>=A['Used']:
            viable_pairs+=1
            
            
```
