```python
def wall_or_space(x,y,constant):
    if x <0 or y<0:
        return 'wall'
    tempval=x*x + 3*x + 2*x*y + y + y*y + constant
    tempbin = bin(tempval)
    numberones=tempbin[2:].count('1')
    if numberones%2==0:
        return 'space'
    elif numberones%2==1:
        return 'wall'
```

```python
def add_tuples(tup1,tup2):
    return (tup1[0]+tup2[0],tup1[1]+tup2[1])
```

```python
# lets create wall_or space dict
constant=1358
wsdict={}
for x in range(-1,100):
    for y in range(-1,100):
        wsdict[(x,y)]=wall_or_space(x,y,constant)
```

```python
class Node():
    def __init__(self,space,parent):
        self.space=space
        self.left=add_tuples(self.space,(-1,0))
        self.left_type=wsdict[self.left]
        self.right=add_tuples(self.space,(1,0))
        self.right_type=wsdict[self.right]
        self.up=add_tuples(self.space,(0,-1))
        self.up_type=wsdict[self.up]
        self.down=add_tuples(self.space,(0,1))
        self.down_type=wsdict[self.down]
        self.children=[]
        self.parent=parent
    
    def create_child(self,loc):
        newnode=Node(loc,self)
        self.children.append(newnode)
        return newnode
        

        
        
```

```python
def search_tree(Node,destination):
    if Node.space==destination:
        return True
    searchval=False
    if len(Node.children)==0:
        searchval=False
    # check to see if value in tree
    for child in Node.children:
        if child.space==destination:
            searchval=True
    if searchval==False:
        for child in Node.children:
            searchval=search_tree(child,destination)
            if searchval==True:
                return True
    return searchval
```
