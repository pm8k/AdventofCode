```python
def parse_df(s):
    strlist=s[0].split(' ')
    for i in range(len(strlist)):
        s[i]=strlist[i]
    return s
```

```python
df=pd.read_csv('Day23.txt',header=None)
df=df.apply(parse_df,axis=1)
i=0
j=0
num_dict={
    'a':7,
    'b':0,
    'c':0,
    'd':0
}
instrs=[]

while i< len(df):
    j+=1
    if j%1000==0:
        print j
    series=df.iloc[i]

    if series[0]=='cpy':
        value=series[1]
        # test to see if instruction is valid - only if second argument is in the dictionary
        value2=series[2]
        if value2 not in num_dict.keys():
            continue
        # if it reaches this point, continue
        if value in num_dict.keys():
            num_dict[series[2]]=num_dict[series[1]]
        else:
            num_dict[series[2]]=int(series[1])
    elif series[0]=='jnz':
        value=series[1]
        if value in num_dict.keys():
            if num_dict[value] != 0:
                i+=int(series[2])
                continue
        else:
            val2=series[2]
            if val2 in num_dict.keys():
                if num_dict[val2]!=0:
                    i+=num_dict[val2]
                    continue
            else:
                if int(series[1])!=0:
                    i+=int(series[2])
                    continue
    elif series[0]=='inc':
        num_dict[series[1]]+=1
    elif series[0]=='dec':
        num_dict[series[1]]-=1
    elif series[0]=='tgl':
        if series[1] in num_dict.keys():
            tempi = i + num_dict[series[1]]
        else:
            tempi = i + int(series[1])
        if tempi >= 0 and tempi < len(df):
            tempseries=df.iloc[tempi]
            #test for one argument instructions
            if pd.isnull(tempseries[2]):
                if tempseries[0]=='inc':
                    tempseries[0]='dec'
                else:
                    tempseries[0]='inc'
            #otherwise its a two argument instruction
            else:
                if tempseries[0]=='jnz':
                    tempseries[0]='cpy'
                else:
                    tempseries[0]='jnz'
    i+=1
                
        

    
```
