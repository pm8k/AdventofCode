```python
def search_tree(Node,destination):
    if Node.space==destination:
        return True
    searchval=False
    if len(Node.children)==0:
        searchval=False
    # check to see if value in tree
    for child in Node.children:
        if child.space==destination:
            searchval=True
    if searchval==False:
        for child in Node.children:
            searchval=search_tree(child,destination)
            if searchval==True:
                return True
    return searchval
```

```python
def add_tuples(tup1,tup2):
    return (tup1[0]+tup2[0],tup1[1]+tup2[1])
```

```python
class Node():
    def __init__(self,space,parent):
        self.space=space
        self.left=add_tuples(self.space,(-1,0))
        self.right=add_tuples(self.space,(1,0))
        self.up=add_tuples(self.space,(0,-1))
        self.down=add_tuples(self.space,(0,1))
        self.children=[]
        self.parent=parent
    
    def create_child(self,loc):
        newnode=Node(loc,self)
        self.children.append(newnode)
        return newnode
```

```python
def find_distance(start_node,end_node):
    firstnode=Node(start_node,None)
    finaldestination=end_node
    childrennodes=[firstnode]
    finalnode=None
    end=False
    j=0
    while not end:
        j+=1
        if j==1000:
            print 'ERRAR'
            return firstnode
        newchildren=[]
        for node in childrennodes:
            #lets add children for our child nodes
            if df.iloc[node.left[1],node.left[0]]!='#':
                #make sure node isn't already in tree
                intree=search_tree(firstnode,node.left)
                if intree==False:
                    newnode=node.create_child(node.left)
                    newchildren.append(newnode)
            if df.iloc[node.right[1],node.right[0]]!='#':
                #make sure node isn't already in tree
                intree=search_tree(firstnode,node.right)
                if intree==False:
                    newnode=node.create_child(node.right)
                    newchildren.append(newnode)
            if df.iloc[node.up[1],node.up[0]]!='#':
                #make sure node isn't already in tree
                intree=search_tree(firstnode,node.up)
                if intree==False:
                    newnode=node.create_child(node.up)
                    newchildren.append(newnode)
            if df.iloc[node.down[1],node.down[0]]!='#':
                #make sure node isn't already in tree
                intree=search_tree(firstnode,node.down)
                if intree==False:
                    newnode=node.create_child(node.down)
                    newchildren.append(newnode)

        childrennodes=newchildren
        for child in childrennodes:
            if child.space==finaldestination:
                finalnode=child
                end=True
        
    i=0
    end=False
    testnode=finalnode.parent
    while not end:
        i+=1
        if testnode==firstnode:
            end=True
            print i
        else:
            testnode=testnode.parent
    return i
```
