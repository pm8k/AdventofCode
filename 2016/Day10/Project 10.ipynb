```python
import pandas as pd
```

```python
df=pd.read_csv('Project10.csv',header=None)
```

```python
def parse_transfers(s):
    split=s[0].split(' ')
    newseries=pd.Series()
    newseries['originbot'] = int(split[1])
    newseries['lowdest'] = split[5]
    newseries['lowvalue'] = int(split[6])
    newseries['highdest'] = split[10]
    newseries['highvalue'] = int(split[11])
    
    return newseries

def parse_values(s):
    split=s[0].split(' ')
    bot=int(split[5])
    chip=int(split[1])
    newseries=pd.Series()
    newseries['bot']=bot
    newseries['chip']=chip
    return newseries
```

```python
values=df[df[0].str.contains('value')]
#values=values.apply(parse_values,axis=1)
transfers = df[~df[0].str.contains('value')]
transfers = transfers.apply(parse_transfers,axis=1)
```

```python
robot_dict = {}
# add values to robot dict
for i,row in values.iterrows():
    split=row[0].split(' ')
    bot=int(split[5])
    chip=int(split[1])
    if bot not in robot_dict.keys():
        robot_dict[bot]=[chip,None]
    else:
        robot_dict[bot][1]=chip
```

```python
chipdf=pd.DataFrame(robot_dict).transpose().reset_index()
chipdf.columns=['Bot','Chip1','Chip2']
chipdf.set_index('Bot',inplace=True)
```
