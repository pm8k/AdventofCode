```python
# Graph Version
# instead of iterating down a single path, create a tree structure searching all paths
# create a numerical configured state and have that s
```

```python
class Node():
    
    def __init__(self,loc_df,elv_loc,parent):
        self.loc_df=loc_df.copy(deep=True)
        self.elv_loc=elv_loc
        self.children=[]
        self.parent=parent
        
    def add_child(self,loc_df,elv_loc):
        newnode=Node(loc_df,elv_loc,self)
        self.children.append(newnode)
        return newnode
        
    
```

```python
def check_floors(df):
    df2=df.drop('elevator',axis=1)
    for i in range(1,5):
        floor=df2.loc[i]
        floor=floor[floor==1]
        if len(floor)>0:
            tempdf= pd.DataFrame([x.split('_') for x in floor.index],columns=['element','type'])
            for element in tempdf['element'].unique():
                subdf=tempdf[tempdf['element']==element]
                if len(subdf)==2:
                    tempdf = tempdf[tempdf['element']!=element]
            if len(tempdf[tempdf['type']=='generator'])>0 and len(tempdf[tempdf['type']=='microchip'])>0:
                return False
        
    return True
```

```python
def loc_possible_moves(df,elv_loc):
    currentdf=df.drop('elevator',axis=1)
    #first get all possible combinations from the current floor
    currfloor=currentdf.loc[elv_loc]
    currfloor=currfloor[currfloor==1].sort_index()
    comblist=currfloor.index.tolist()
    comblist2=[[x] for x in comblist]

    comblist2+=[list(x) for x in itertools.combinations(comblist,2)]
    
    if elv_loc==1:
        comblist2=[['up']+x for x in comblist2]
    elif elv_loc==4:
        comblist2=[['down']+x for x in comblist2]
    else:
        comblist2=[['down']+x for x in comblist2]+[['up']+x for x in comblist2]

    
    return comblist2

def create_temp_dfs(comblist,df,elv_loc):
    dflist=[]
    for comb in comblist:
        df2=df.copy(deep=True)
        direc=comb[0]
        if direc=='up':
            val=1
        elif direc=='down':
            val=-1
        items=comb[1:]
        for item in items:
            df2.loc[elv_loc,item]=0
            df2.loc[elv_loc+val,item]=1
            df2.loc[elv_loc,'elevator']=0
            df2.loc[elv_loc+val,'elevator']=1
            
        dflist.append(df2)
            
    return dflist

# iterate over each floor to see if the current configuration is valid
def check_floors(df):
    df2=df.drop('elevator',axis=1)
    for i in range(1,5):
        floor=df2.loc[i]
        floor=floor[floor==1]
        if len(floor)>0:
            tempdf= pd.DataFrame([x.split('_') for x in floor.index],columns=['element','type'])
            for element in tempdf['element'].unique():
                subdf=tempdf[tempdf['element']==element]
                if len(subdf)==2:
                    tempdf = tempdf[~((tempdf['element']==element)&(tempdf['type']=='microchip'))]
#                     print '\n\n',element
#                     display(tempdf)
#                     print '\n'
            if len(tempdf[tempdf['type']=='generator'])>0 and len(tempdf[tempdf['type']=='microchip'])>0:
                return False
        #print tempdf
    return True

def search_tree(Node,df):
    # check if node has the value we want
    # if so, return true
    if Node.loc_df.equals(df):
        return True
    # we haven't found it yet, so the searchval is False
    searchval=False
    # if there are no children, it remains false, and is returned
    if len(Node.children)==0:
        searchval=False
    # if there are children, iterate over them, calling function recursively
    else:
        for child in Node.children:
            searchval=search_tree(child,df)
            # if a child is the target value, then we can stop iterating and return
            if searchval==True:
                return True
    # if it reaches this point, it will return False
    return searchval
```
