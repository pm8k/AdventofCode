```python
def rotate(t,r):
    rotate_order = [(0,1),(-1,0),(0,-1),(1,0)]
    if t == 'L':
        new_dir = rotate_order[(rotate_order.index(r)+1)%4]
    else:
        new_dir = rotate_order[(rotate_order.index(r)-1)%4]
    return new_dir
def add_tuples(a,b):
    return (a[0]+b[0],a[1]+b[1])
def mult_tuple(a,c):
    return (a[0]*c,a[1]*c)
```
