```python
def parse_text(s):
    textlist=s[0].split(' ')
    s=pd.Series()
    if textlist[0]=='swap':
        s['action']='swap'
        try:
            s['swaptype']='int'
            s['swapx']=int(textlist[2])
            s['swapy']=int(textlist[5])
        except:
            s['swaptype']='char'
            s['swapx']=textlist[2]
            s['swapy']=textlist[5]
    elif textlist[0]=='rotate':
        s['action']='rotate'
        if textlist[1]=='based':
            s['rotatetype']='char'
            s['rotateletter']=textlist[6]
        else:
            s['rotatetype']='dir'
            s['rotatedir']=textlist[1]
            s['rotatenum']=textlist[2]
    elif textlist[0]=='reverse':
        s['action']='reverse'
        s['reversex']=int(textlist[2])
        s['reversey']=int(textlist[4])
    elif textlist[0]=='move':
        s['action']='move'
        s['movex']=int(textlist[2])
        s['movey']=int(textlist[5])
    return s
```

```python
outdf=df.apply(parse_text,axis=1)
```

```python
def rotate_string(charlist,rotdir):
    if rotdir=='right':
        list1=charlist[:-1]
        list2=charlist[-1:]
        charlist=list2+list1
    elif rotdir=='left':
        list1=charlist[:1]
        list2=charlist[1:]
        charlist=list2+list1
    return charlist
```

```python
def run_action(s,charlist):
    if s.action=='swap':
        if s.swaptype=='char':
            charx=s.swapx
            chary=s.swapy
            xind=charlist.index(charx)
            yind=charlist.index(chary)
            charlist[xind]=chary
            charlist[yind]=charx
        elif s.swaptype=='int':
            xind=s.swapx
            yind=s.swapy
            charx=charlist[xind]
            chary=charlist[yind]
            charlist[xind]=chary
            charlist[yind]=charx
    elif s.action=='rotate':
        if s.rotatetype=='char':
            index=charlist.index(s.rotateletter)
            rotatenum=1+index
            if index>=4:
                rotatenum+=1
            for i in range(rotatenum):
                charlist=rotate_string(charlist,'right')
#             list1=charlist[:-rotatenum]
#             list2=charlist[-rotatenum:]
#             charlist=list2+list1
        if s.rotatetype=='dir':
            rotatenum=int(s.rotatenum)
            for i in range(rotatenum):
                charlist=rotate_string(charlist,s.rotatedir)
#             if s.rotatedir=='right':
#                 rotatenum=int(s.rotatenum)

#                 list1=charlist[:-rotatenum]
#                 list2=charlist[-rotatenum:]
#                 charlist=list2+list1
#             elif s.rotatedir=='left':
#                 rotatenum=int(s.rotatenum)

#                 list1=charlist[:rotatenum]
#                 list2=charlist[rotatenum:]
#                 charlist=list2+list1
    elif s.action=='reverse':
        rx=int(s.reversex)
        ry=int(s.reversey)
        l1=charlist[:rx]
        l2=charlist[rx:ry+1][::-1]
        l3=charlist[ry+1:]
        charlist=l1+l2+l3
    elif s.action=='move':
        movex=int(s.movex)
        movey=int(s.movey)
        char=charlist.pop(movex)
        charlist.insert(movey,char)
    return charlist
```
